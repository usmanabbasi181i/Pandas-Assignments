```python
import numpy as np
import pandas as pd
import glob
```

```python
all_files=glob.glob("states[0-9].csv")
```

```python
li=[]
for filename in all_files:
    df=pd.read_csv(filename,index_col=None,header=0)
    li.append(df)
```

```python
us_census=pd.concat(li,axis=0,ignore_index=True)
us_census
```

```python
us_census.head(10)
```

```python
us_census.columns
```

```python
us_census.dtypes
```

```python
us_census.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
us_census[["Males","Females"]]=us_census.GenderPop.str.split("_",expand=True)
```

```python
us_census.drop(columns=["GenderPop"],inplace=True)
```

```python
us_census.head(3)
```

```python
us_census["Income"]=us_census.Income.str.strip("$")
us_census=us_census.replace("%","",regex=True)
```

```python
us_census=us_census.replace("M","",regex=True)
```

```python
us_census=us_census.replace("F","",regex=True)
```

```python
us_census.head(3)
```

```python
us_census.loc[:,"Hispanic":"Income"]=round(us_census.loc[:,"Hispanic":"Income"].apply(pd.to_numeric),2)
```

```python
us_census.head(3)
```

```python
us_census.dtypes
```

```python
us_census["Males"]=us_census["Males"].astype(int)
```

```python
us_census.dtypes
```

```python
us_census["Females"][us_census["Females"]==""]
```

```python
us_census.drop(columns=["Females"],inplace=True)
```

```python
us_census["Female"]=us_census["TotalPop"]-us_census["Males"]
```

```python
us_census["Female"]=us_census["Female"].astype(int)
```

```python
us_census.dtypes
```

```python
us_census.duplicated().any()
us_census.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.scatter(us_census["Female"],us_census["Income"])
plt.xlabel("Income/Salary $",fontsize=18)
plt.ylabel("Female_Population",fontsize=15)
plt.show()
```
